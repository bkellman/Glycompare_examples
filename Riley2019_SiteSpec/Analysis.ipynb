```r
library(gplots)
library("viridis")   
```

```r
r_comp = t(read.csv('data-in/paper_nature_abundance_table.csv',row.names = 1))
r_subt = t(t(read.csv('data-in/paper_nature_motif_abd_table_composition.csv',row.names = 1)))
head(r_comp)
head(r_subt)
```

```r
dim(r_comp)
dim(r_subt)
```

```r

pdf('results/heatmaps.sel.pdf',height=20,width=20)

##### select proteins

# Q91ZX7|Q6P6I8 : LIRP1(Q91ZX7) & SIRPA(Q6P6I8) & LAMP1 (P11438)
# P11627|Q810U4 : L1CAM (P11627) & NRCAM (Q810U4)
ccol=c('red','black','grey')[factor(gsub('_.*','',colnames(r_comp)[idx]))]
idx = grepl('Q91ZX7|Q6P6I8|P11438',colnames(r_comp))
heatmap.2(log2(unique(r_comp[,idx])+1),trace='none',col=inferno,ColSideColors = ccol,main='Composition Abundance\nLIRP1(Q91ZX7) & SIRPA(Q6P6I8)& LAMP1 (P11438)',mar=c(10,10))
heatmap.2(log2(unique(r_subt[,idx])+1),trace='none',col=inferno,ColSideColors = ccol,main='Substructure Abundance\nLIRP1(Q91ZX7) & SIRPA(Q6P6I8)& LAMP1 (P11438)',mar=c(10,10))
       
cr_comp=(cor(log2(unique(r_comp[,idx])+1),method='spearman'))^2
cr_subt=(cor(log2(unique(r_subt[,idx])+1),method='spearman'))^2
heatmap.2(cr_comp,trace='none',col=plasma,breaks=seq(0,1,.01), ColSideColors = ccol,main='Composition Abundance Spearman R^2\nLIRP1(Q91ZX7) & SIRPA(Q6P6I8)& LAMP1 (P11438)',mar=c(10,10))
heatmap.2(cr_subt,trace='none',col=plasma,breaks=seq(0,1,.01),ColSideColors = ccol,main='Substructure Abundance Spearman R^2\nLIRP1(Q91ZX7) & SIRPA(Q6P6I8)& LAMP1 (P11438)',mar=c(10,10))
dev.off()

pdf('results/heatmaps.all.pdf',height=20,width=20)
###  all proteins
heatmap.2(log(unique(r_comp)+1),trace='none',col=inferno,main='Counts: Composition Abundance')
heatmap.2(log(unique(r_subt)+1),trace='none',col=inferno,main='Counts: Substructure Abundance')
       
cr_comp=(cor(log(unique(r_comp)+1),method='spearman'))^2
cr_subt=(cor(log(unique(r_subt)+1),method='spearman'))^2
heatmap.2(cr_comp,trace='none',col=plasma,breaks=seq(0,1,.01),main='Counts: Composition Abundance Spearman R^2')
heatmap.2(cr_subt,trace='none',col=plasma,breaks=seq(0,1,.01),main='Counts: Substructure Abundance Spearman R^2')

dev.off()
```
